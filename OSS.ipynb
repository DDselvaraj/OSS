```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_excel('/content/bmi.xlsx')
df.head()
```

```python
table=pd.pivot_table(data=df,index=['BMI','Gender'],values='Status')
table
```

```python
df.memory_usage(deep=True)
```

```python
df['Status'].value_counts()
```

```python
import matplotlib.pyplot as plt
import numpy as np
height = np.asarray(df.Height)
weight = np.asarray(df.Weight)
bmi= np.asarray(df.BMI)
plt.scatter(x=height, y=weight, s=bmi * 2)
plt.xlabel('Height')
plt.ylabel('Weight')
plt.title('Height Vs Weight')
plt.show()
```

```python
print('The mean of weights in the dataset is:',df['Weight'].mean())
print('The maximum BMI in the dataset is:',df['BMI'].max())
print('The minimum BMI in the dataset is:',df['BMI'].min())
print('The total sum of BMI in the dataset is:',df['BMI'].sum())
```

```python
plt.bar(df.Status,df.BMI,color='maroon',width=0.5)
plt.xlabel('Status')
plt.ylabel('BMI')
plt.title('BMI of people w.r.t Status')
plt.show()
```

```python
plt.scatter(df.Status,df.BMI,color= "green",marker= "*", s=30)
plt.xlabel('Status')
plt.ylabel('BMI')
plt.title('BMI of people w.r.t Status')
```

```python
plt.plot(df.Height,df.Weight, "o")
plt.xlabel('Height')
plt.ylabel('Weight')
plt.title('Height vs Weight Graph')
plt.show()
```
